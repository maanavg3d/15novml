```python
def age_treat(x):
    age = x['Age']
    sex = x['Sex']
    pclass = x['Pclass']
    if pd.isnull(age):
        return df['Age'][(df['Sex']==sex) & (df['Pclass']==pclass)].mean()
    return age
```

```python
df['Age'] = df[['Sex','Pclass','Age']].apply(age_treat, axis=1)
```

```python
df.dropna(axis=0,inplace=True)
```
