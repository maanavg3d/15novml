```python
df = pd.get_dummies(df)
```

```python
del df['CustomerID']
```

```python
from sklearn.impute import KNNImputer
```

```python
df1 = pd.DataFrame(KNNImputer().fit_transform(df),
                   columns=df.columns)
```

```python
pd.isnull(df1).sum()*100/len(df1)
```

```python
x = df1.drop('ProdTaken',axis=1)
y = df1['ProdTaken']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,
                                             random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier().fit(xtrain,ytrain)
```

```python
model_dt.classes_
```

```python
model_dt.feature_importances_
```

```python
imp = pd.Series(model_dt.feature_importances_,index=x.columns).sort_values()
plt.figure(figsize=(4,12))
imp.plot.barh()
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
# Building the Tree
plt.figure(figsize=(15,18))
f = plot_tree(model_dt, feature_names=x.columns,
             filled=True, fontsize=9, node_ids=True,
             class_names=True)
for i in f:
    arrow = i.arrow_patch
    if arrow is not None:
        arrow.set_edgecolor('black')
        arrow.set_linewidth(1)
plt.show()
```

```python
print('Training Score:',100*model_dt.score(xtrain,ytrain))
print('Test Score:',100*model_dt.score(xtest,ytest))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier().fit(xtrain,ytrain)
```

```python
print('Training Score:',100*model_rf.score(xtrain,ytrain))
print('Test Score:',100*model_rf.score(xtest,ytest))
```

## GRID SEARCH CV

```python
from sklearn.model_selection import GridSearchCV, KFold
```

```python
params_dt = {'criterion':['gini','entropy'],
             'max_depth':[1,2,3,4,5,6,None],
             'min_samples_split':[2,4,6,8,10],
             
            }
```

```python
from datetime import datetime as dt
```

```python
print(dt.now())
model_grid = GridSearchCV(DecisionTreeClassifier(),
                          param_grid=params_dt, verbose=3,
                     ).fit(xtrain,ytrain)
print(dt.now())
```

```python
model_grid.best_params_
```

```python
model_grid.best_score_
```

```python
model_grid_dt = model_grid.best_estimator_
```

```python
print('Training Score:',100*model_grid_dt.score(xtrain,ytrain))
print('Test Score:',100*model_grid_dt.score(xtest,ytest))
```

```python
params_rf = {
    'n_estimators':[20,40,60,80,100,150,200],
    'max_depth':[10,20,40,None],
    'min_samples_leaf':[5,10,15,20],
    'min_samples_split':[10,20,40,80,100],
}
```

```python
print(dt.now())
rf_grid = GridSearchCV(RandomForestClassifier(), cv=10,
                       param_grid=params_rf, verbose = 3,
                      ).fit(xtrain,ytrain)
print(dt.now())
```
